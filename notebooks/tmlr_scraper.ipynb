```python
import requests
import csv
import PyPDF2
from bs4 import BeautifulSoup
import pandas as pd
import io
import os
```

```python

# Base URL of the site to scrape
base_url = "https://jmlr.org/tmlr/papers/"

# Get the HTML content of the page
response = requests.get(base_url)
response.raise_for_status()

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all paper links
paper_links = list(set(  # Remove duplicates
    link['href'] for link in soup.find_all('a', href=True)
    if link['href'].startswith("https://openreview.net/pdf?id=")
))

# Function to extract title and abstract from a PDF
def extract_pdf_data(pdf_url):
    try:
        # Fetch the PDF content
        pdf_response = requests.get(pdf_url)
        pdf_response.raise_for_status()

        # Read the PDF content
        pdf_file = io.BytesIO(pdf_response.content)
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        # Combine all pages' text
        text = " ".join(page.extract_text() for page in pdf_reader.pages if page.extract_text())

        # Extract title (skip metadata and find the first significant line)
        lines = [line.strip() for line in text.split("\n") if line.strip()]
        title = None
        for line in lines:
            if not line.startswith("Published in Transactions on Machine Learning Research"):  # Skip metadata lines
                title = line
                break

        if not title:
            title = "Title not found"

        # Extract abstract (search for the keyword 'abstract')
        abstract_start = text.lower().find('abstract')
        if abstract_start != -1:
            abstract_end = text.lower().find('introduction', abstract_start)
            abstract = text[abstract_start + 8:abstract_end].strip() if abstract_end != -1 else text[abstract_start + 8:].strip()
        else:
            abstract = "Abstract not found"

        return title, abstract
    except Exception as e:
        return "Error extracting title", f"Error extracting abstract: {str(e)}"

# Store data in a list
data = []

# Loop through the paper links
for i, paper_url in enumerate(paper_links[:600], start=1):  # Limit to 100 papers
    print(f"Processing paper {i}: {paper_url}")
    title, abstract = extract_pdf_data(paper_url)
    data.append({
        'title': title,
        'abstract': abstract,
        'URL': paper_url
    })
```

```python
# Create a DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv('tmlr_500.csv', index=False, escapechar='\\')

print("Data saved to papers_final.csv")
```
